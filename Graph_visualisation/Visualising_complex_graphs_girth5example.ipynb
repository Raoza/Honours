```python
#Plotting the graph


#Define the orbits, label doesn't necessarily matter can view all 6 arrangements of the orbits and choose the cleanest one
orbits = {
    'large_cycle': [0, 3, 13, 17, 14, 21, 19, 9, 16, 18, 7, 1],
    'medium_cycle': [2, 8, 5, 11, 20, 10],
    'disjoint': [(6, 12), (15, 4)]
}

# Initial plot settings
pos = {}
radius_large = 3  # Outer cycle
radius_medium = 2  # Inner cycle
radius_disjoint = 1  # Central for disjoint components

# Assign positions for the large cycle
angle_step = 2*pi/len(orbits['large_cycle'])
for i, vertex in enumerate(orbits['large_cycle']):
    angle = i * angle_step
    pos[vertex] = (radius_large*cos(angle), radius_large*sin(angle))

# Assign positions for the medium cycle with a -40 degrees rotation
rotation_degrees = -75
rotation_radians = rotation_degrees * (pi / 180)  # Convert degrees to radians
angle_step = 2*pi/len(orbits['medium_cycle'])
for i, vertex in enumerate(orbits['medium_cycle']):
    angle = i * angle_step + rotation_radians
    pos[vertex] = (radius_medium*cos(angle), radius_medium*sin(angle))

# Assign positions for disjoint components (place in center with slight offsets)
pos[6] = (radius_disjoint*cos(pi/4), radius_disjoint*sin(pi/4))
pos[12] = (radius_disjoint*cos(5*pi/4), radius_disjoint*sin(5*pi/4))
pos[15] = (radius_disjoint*cos(3*pi/4), radius_disjoint*sin(3*pi/4))
pos[4] = (radius_disjoint*cos(7*pi/4), radius_disjoint*sin(7*pi/4))

# Plot the graph
graph_plotmax22 = maxg22.plot(pos=pos, vertex_size=60, vertex_labels=False).show()
#graph_plotmax22.save('maxg22.png')
```
